```python
import argparse
import os
import os.path as osp

from mmengine.config import Config, DictAction
from mmengine.logging import MMLogger
from mmengine.runner import Runner
from mmedit.utils import register_all_modules
register_all_modules(init_default_scope=False)
# build the runner from config
cfg = Config.fromfile('configs/mattors/gca/baseline_r34_4x10_200k_comp1k.py')
cfg.work_dir=""
runner = Runner.from_cfg(cfg)
```

```python
runner.optim_wrapper = runner.build_optim_wrapper(runner.optim_wrapper)
runner.param_schedulers = runner.build_param_scheduler(runner.param_schedulers)  # type: ignore
```

```python
from mmengine import ParamSchedulerHook 
for hook in runner.hooks:
    if isinstance(hook, ParamSchedulerHook):
        break
assert isinstance(hook, ParamSchedulerHook)
```

```python
for i in range(cfg.train_cfg.max_iters):
    lr = runner.optim_wrapper.get_lr()
    if i ==0:
        print(lr)
    if (i+1) % 1000 == 0:
        print(i+1, lr)
    hook.after_train_iter(runner, None)
```
